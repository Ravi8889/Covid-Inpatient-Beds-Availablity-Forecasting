```python
#Smoothing techniques
from statsmodels.tsa.stattools import adfuller
x=df['Total Inpatient Beds'].values
result= adfuller(x)
#print(result)
print('ADF Statistic: %.3f' % result[0])
print('p-value: %3.f'% result[1])
print("Lags- Used: %.3f" % result[2])
print('Critical Values:')
for key,value in result[4].items():
    print('\t%s: %.3f' % (key, value))
#The Intuation of Dicky Fuller Test 
#Here the Critical value is Less than 10% , thus the Time Series is Stationary with the 90% confidence
```

```python
# here we use auto correlation and Partial Auto Correlation
#ARIMA has 3 componets they are 1 1AutoRegressive, 2. Integration  3. Moving Average
#indetification of AR model using ACF plot  in ARIMA we need 3 values p,d,q p == ar model lags 
#d = d is diffrencing and q ==> q is autoregrssive
```

```python
from statsmodels.tsa.stattools import acf, pacf
fig = plt.figure(figsize=(12,8))
##ACF plots 
from statsmodels.graphics.tsaplots import plot_acf 
rk2 =fig.add_subplot(211)
fig =plot_acf(df['Total Inpatient Beds'],lags=40,ax =rk2)
#plt.show()

#PACF Plots
from statsmodels.graphics.tsaplots import plot_pacf
#df= pd.read_csv(r"C:\Users\RaviKiran\OneDrive\Desktop\Data Science\Data Sets\daily-minimum-temperatures.csv", header=0, index_col=0)
rk1=fig.add_subplot(212)
fig =plot_pacf(df['Total Inpatient Beds'],lags= 40,ax =rk1)
#rk1=fig.add_subplot(212)# -1 to +1
#plt.show()
# here ACF has exponentatial Decrease with here q -value will be 4 or 5
# Pacf is haves the sudden shut's off at 3 so the ar value is 3 here the p-value will be 3 or 2
```

```python

# For non-seasonal data
#p=1, d=1, q=0 or 1
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(df['Total Inpatient Beds'],order=(2,1,5))
model_fit=model.fit()
```

```python
model_fit.summary()
```

```python
df['forecast']=model_fit.predict(start=300,end=358,dynamic=True)
df[['Total Inpatient Beds','forecast']].plot(figsize=(12,8))
```
